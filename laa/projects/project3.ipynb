```python
import numpy as np
import matplotlib.pyplot as plt
import imageio.v2 as imageio
from PIL import Image

# 1. Load the color image
ImJPG = imageio.imread('daylilies.jpg')
plt.imshow(ImJPG)
plt.title('Original Image')
plt.axis('off')
plt.show()
```

```python
# 2. Get image dimensions
m, n, l = ImJPG.shape
print(f'Image Dimensions: {m}x{n}x{l}')
```

```python
# 3. Extract color channels
redChannel = ImJPG[:, :, 0]
greenChannel = ImJPG[:, :, 1]
blueChannel = ImJPG[:, :, 2]

plt.figure(figsize=(10, 3))
plt.subplot(1, 3, 1)
plt.imshow(redChannel, cmap='gray')
plt.title('Red Channel')
plt.axis('off')
plt.subplot(1, 3, 2)
plt.imshow(greenChannel, cmap='gray')
plt.title('Green Channel')
plt.axis('off')
plt.subplot(1, 3, 3)
plt.imshow(blueChannel, cmap='gray')
plt.title('Blue Channel')
plt.axis('off')
plt.show()
```

```python
# 4. Convert image to grayscale using a weighted sum filter
GrayMatrix = np.array([[1/3, 1/3, 1/3], [1/3, 1/3, 1/3], [1/3, 1/3, 1/3]])
ImJPG_Gray = np.dot(ImJPG[..., :3], GrayMatrix.T).astype(np.uint8)
plt.imshow(ImJPG_Gray, cmap='gray')
plt.title('Grayscale Image')
plt.axis('off')
plt.show()
```

```python
# 5. Apply a sepia filter
SepiaMatrix = np.array([[0.393, 0.769, 0.189], [0.349, 0.686, 0.168], [0.272, 0.534, 0.131]])
ImJPG_Sepia = np.dot(ImJPG[..., :3], SepiaMatrix.T)
ImJPG_Sepia = np.clip(ImJPG_Sepia, 0, 255).astype(np.uint8)
plt.imshow(ImJPG_Sepia)
plt.title('Sepia Image')
plt.axis('off')
plt.show()
```

```python
# 6. Apply a red filter
RedMatrix = np.array([[1, 0, 0], [0, 0, 0], [0, 0, 0]])
ImJPG_Red = np.dot(ImJPG[..., :3], RedMatrix.T).astype(np.uint8)
plt.imshow(ImJPG_Red)
plt.title('Red Filtered Image')
plt.axis('off')
plt.show()
```

```python
# 7. Permute colors (swap R, G, B)
PermuteMatrix = np.array([[0, 0, 1], [1, 0, 0], [0, 1, 0]])
ImJPG_Permute = np.dot(ImJPG[..., :3], PermuteMatrix.T).astype(np.uint8)
plt.imshow(ImJPG_Permute)
plt.title('Permuted Colors Image')
plt.axis('off')
plt.show()
```

```python
# 8. Remove green color
DeleteGreenMatrix = np.array([[1, 0, 0], [0, 0, 0], [0, 0, 1]])
ImJPG_DeleteGreen = np.dot(ImJPG[..., :3], DeleteGreenMatrix.T).astype(np.uint8)
plt.imshow(ImJPG_DeleteGreen)
plt.title('Image without Green Channel')
plt.axis('off')
plt.show()
```

```python
# 9. Invert colors
ImJPG_Invert = 255 - ImJPG
plt.imshow(ImJPG_Invert)
plt.title('Inverted Colors Image')
plt.axis('off')
plt.show()
```

```python
# 10. Increase contrast
ImJPG_HighContrast = np.clip(ImJPG * 1.5, 0, 255).astype(np.uint8)
plt.imshow(ImJPG_HighContrast)
plt.title('High Contrast Image')
plt.axis('off')
plt.show()
```

```python
# 11. Reduce contrast
ImJPG_LowContrast = np.clip(ImJPG * 0.5, 0, 255).astype(np.uint8)
plt.imshow(ImJPG_LowContrast)
plt.title('Low Contrast Image')
plt.axis('off')
plt.show()
```

```python
# 12. Convert to black and white
ImJPG_BW = np.where(ImJPG.mean(axis=2) > 128, 255, 0).astype(np.uint8)
plt.imshow(ImJPG_BW, cmap='gray')
plt.title('Black and White Image')
plt.axis('off')
plt.show()
```

```python
# 13. Reduce image to 8 shades
ImJPG_8Shades = (ImJPG // 32) * 32
plt.imshow(ImJPG_8Shades)
plt.title('8 Shades Image')
plt.axis('off')
plt.show()
```

```python
# 14. Apply gamma correction
gamma = 0.5
ImJPG_Gamma = np.clip((ImJPG / 255.0) ** gamma * 255, 0, 255).astype(np.uint8)
plt.imshow(ImJPG_Gamma)
plt.title('Gamma Corrected Image')
plt.axis('off')
plt.show()
print("Project 3 Image Processing Complete")
```
